```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ["EIN", "NAME"])
```

```python
application_df.head() # (To verify that the columns are not there anymore)
```

```python
# Determine the number of unique values in each column. (The results are the same as the example output before)
for i in application_df.columns:
    print(i, len(application_df[i].unique())) # (The values' "dtype" is int64, and that information was given to us)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_typecount = application_df["APPLICATION_TYPE"].value_counts()
application_typecount
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace` (At random, I chose the cutoff value to be less than 200)
application_replace = list(application_typecount[application_typecount<200].index)
application_replace

# Replace in dataframe
for app in application_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_class = application_df["CLASSIFICATION"].value_counts()
```

```python
application_class
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_1 = application_class[application_class>1]
```

```python
class_1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace` (At random I choose the value of 100 as the cutoff value)
classifications_replace = list(class_1[class_1<100].index)

# Replace in dataframe
for cls in classifications_replace:
    application_df["CLASSIFICATION"] = application_df["CLASSIFICATION"].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df["CLASSIFICATION"].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df,dtype=float)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
x = application_df.drop("IS_SUCCESSFUL", axis=1).values # (Those values have to be dropped for future calculations, despite
                                                        # needing to call it just before)
    
# Split the preprocessed data into a training and testing dataset (Most of this is based off the example code later)
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Attempt 1 (Mutliples of 5 for hidden node layers)

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=5
hidden_nodes_layer2=10
hidden_nodes_layer3=15
nn = tf.keras.models.Sequential()

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model (A lot of this was done in class)
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
model_fit = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Observation: Appears to be more accurate than the Starter Code

# Attempt 3 (Adding more hidden layers, changing multiples, and activation)

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer. (Multiples of 10)
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=10
hidden_nodes_layer2=20
hidden_nodes_layer3=30
hidden_nodes_layer4=40
hidden_nodes_layer5=50
nn = tf.keras.models.Sequential()

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Third Hidden Layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation='relu'))

# Fourth Hidden Layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='relu'))

# Check the structure of the model
nn.summary()
```
